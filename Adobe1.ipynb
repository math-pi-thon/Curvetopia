```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def read_csv_(csv_path):
    np_path_XYs = np.genfromtxt(csv_path, delimiter=',')
    path_XYs = []
    for i in np.unique(np_path_XYs[:, 0]):
        npXYs = np_path_XYs[np_path_XYs[:, 0] == i][:, 1:]
        XYs = []
        for j in np.unique(npXYs[:, 0]):
            XY = npXYs[npXYs[:, 0] == j][:, 1:]
            XYs.append(XY)
        path_XYs.append(XYs)
    return path_XYs

def plot(paths_XYs, title, ax):
    colours = ['red', 'green', 'blue', 'yellow', 'purple']  # Define some colors for plotting
    for i, XYs in enumerate(paths_XYs):
        c = colours[i % len(colours)]
        for XY in XYs:
            ax.plot(XY[:, 0], XY[:, 1], c=c, linewidth=2)
    ax.set_aspect('equal')
    ax.set_title(title)

def convert_to_contour(paths_XYs):
    contours = []
    for XYs in paths_XYs:
        for XY in XYs:
            contour = np.array(XY, dtype=np.int32).reshape((-1, 1, 2))
            contours.append(contour)
    return contours

def smooth_contours(contours, epsilon_factor=0.01):
    smoothed_contours = []
    for contour in contours:
        epsilon = epsilon_factor * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)
        smoothed_contours.append(approx)
    return smoothed_contours

def convert_contours_to_paths(smoothed_contours):
    smoothed_paths = []
    for contour in smoothed_contours:
        path = contour.reshape((-1, 2)).tolist()
        smoothed_paths.append(path)
    return smoothed_paths

def process_and_smooth_paths(csv_path):
    paths_XYs = read_csv_(csv_path)
    contours = convert_to_contour(paths_XYs)
    smoothed_contours = smooth_contours(contours)
    smoothed_paths = convert_contours_to_paths(smoothed_contours)
    return smoothed_paths

def format_for_plotting(smoothed_paths):
    formatted_paths = []
    for path in smoothed_paths:
        formatted_paths.append([np.array(path)])
    return formatted_paths

# Read and process the CSV files
csv_path1 = "frag2.csv"
csv_path2 = "isolated_sol.csv"

output_data1 = read_csv_(csv_path1)
output_data2 = read_csv_(csv_path2)

smoothed_data1 = process_and_smooth_paths(csv_path1)
smoothed_data2 = process_and_smooth_paths(csv_path2)

formatted_smoothed_data1 = format_for_plotting(smoothed_data1)
formatted_smoothed_data2 = format_for_plotting(smoothed_data2)

# Plot side by side
fig, axs = plt.subplots(1, 2, tight_layout=True, figsize=(16, 8))
plot(output_data1, 'Original Data', axs[0])
plot(formatted_smoothed_data1, 'Smoothed Data 1', axs[1])

fig, axs = plt.subplots(1, 2, tight_layout=True, figsize=(16, 8))
plot(output_data2, 'Original Data', axs[0])
plot(formatted_smoothed_data2, 'Smoothed Data 2', axs[1])

plt.show()
```

```python
def identify_shape(polyline):
  num_vertices = len(polyline)
  angles = []
  distance_ratios = []

  for i in range(num_vertices):
    v1 = polyline[i]
    v2 = polyline[(i + 1) % num_vertices]
    v3 = polyline[(i + 2) % num_vertices]

    # Angle calculation (using dot product)
    vec1 = np.subtract(v2, v1)
    vec2 = np.subtract(v3, v2)
    angle = np.arccos(np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2)))
    angles.append(angle)

    # Distance ratio calculation
    dist1 = np.linalg.norm(vec1)
    dist2 = np.linalg.norm(vec2)
    if dist1 != 0:
      distance_ratios.append(dist2 / dist1)

  # Basic shape classification based on number of vertices
  if num_vertices == 3:
    shape = "Triangle"
  elif num_vertices == 4:
    # Further classification based on angles and distance ratios
    if all(np.isclose(angle, np.pi/2) for angle in angles) and all(np.isclose(ratio, 1) for ratio in distance_ratios):
      shape = "Square"
    elif all(np.isclose(angle, np.pi/2) for angle in angles):
      shape = "Rectangle"
    else:
      shape = "Quadrilateral"
  # ... (Add more shape classifications as needed)
  else:
    shape = "Unknown"

  return shape
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_csv_(csv_path):
    np_path_XYs = np.genfromtxt(csv_path, delimiter=',')
    path_XYs = []
    for i in np.unique(np_path_XYs[:, 0]):
        npXYs = np_path_XYs[np_path_XYs[:, 0] == i][:, 1:]
        XYs = []
        for j in np.unique(npXYs[:, 0]):
            XY = npXYs[npXYs[:, 0] == j][:, 1:]
            XYs.append(XY)
        path_XYs.append(XYs)
    return path_XYs

def interpolate_missing_points(XY1, XY2, num_points=10):
    x_vals = np.linspace(XY1[-1, 0], XY2[0, 0], num=num_points)
    y_vals = np.linspace(XY1[-1, 1], XY2[0, 1], num=num_points)
    return np.vstack((x_vals, y_vals)).T

def complete_polylines(paths_XYs):
    completed_paths = []
    for shape in paths_XYs:
        completed_shape = []
        for i in range(len(shape) - 1):
            completed_shape.append(shape[i])
            interp_points = interpolate_missing_points(shape[i], shape[i + 1])
            completed_shape.append(interp_points)
        completed_shape.append(shape[-1])
        completed_paths.append(completed_shape)
    return completed_paths

def plot(paths_XYs, title, ax):
    colours = ['red', 'green', 'blue', 'yellow', 'purple']
    for i, XYs in enumerate(paths_XYs):
        c = colours[i % len(colours)]
        for XY in XYs:
            ax.plot(XY[:, 0], XY[:, 1], c=c, linewidth=2)
    ax.set_aspect('equal')
    ax.set_title(title)

# Read and process the CSV files
csv_path1 = "frag2.csv"
csv_path2 = "frag01_sol.csv"

output_data1 = read_csv_(csv_path1)
output_data2 = read_csv_(csv_path2)

# Complete the polylines
completed_data = complete_polylines(output_data1)

# Plot side by side
fig, axs = plt.subplots(1, 2, tight_layout=True, figsize=(16, 8))
plot(output_data1, 'Original Data', axs[0])
plot(completed_data, 'Completed Data', axs[1])
plt.show()

# Function to generate CSV output from completed polylines
def generate_csv_output(completed_paths, output_path):
    all_data = []
    shape_id = 0
    for shape in completed_paths:
        for line_id, line in enumerate(shape):
            for point in line:
                all_data.append([shape_id, line_id, point[0], point[1]])
        shape_id += 1
    np.savetxt(output_path, all_data, delimiter=',', fmt='%d,%d,%.2f,%.2f')

# Save completed polylines to a new CSV file
generate_csv_output(completed_data, "completed_polylines.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate gradient changes
def calculate_gradients(points):
    gradients = np.gradient(points, axis=0)
    return gradients

# Function to identify sharp angles
def identify_sharp_angles(gradients, threshold=0.5):
    # Calculate change in gradients
    delta_gradients = np.diff(gradients, axis=0)
    # Calculate the magnitude of the changes
    magnitude = np.linalg.norm(delta_gradients, axis=1)
    # Identify sharp changes
    sharp_points = np.where(magnitude > threshold)[0] + 1
    # Return an empty array if no sharp points are found to handle the edge case
    return sharp_points if sharp_points.size else np.array([], dtype=int)

# Function to create polylines
def create_polylines(points, sharp_points):
    polylines = []
    if sharp_points.size: # Only proceed if sharp points were detected
        start = sharp_points[0]
        for end in sharp_points[1:]:
            polyline = points[start:end + 1]
            polylines.append(polyline)
            start = end
    return polylines

# Example usage
points = np.array([[0, 0], [1, 2], [2, 4], [3, 2], [4, 0], [5, -2], [6, -4], [7, -2], [8, 0]])
gradients = calculate_gradients(points)
sharp_points = identify_sharp_angles(gradients, threshold=2)
polylines = create_polylines(points, sharp_points)

# Plot the results
for polyline in polylines:
    plt.plot(polyline[:, 0], polyline[:, 1])

plt.scatter(points[:, 0], points[:, 1], color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def read_csv(csv_path):
    data = np.genfromtxt(csv_path, delimiter=',')
    polylines = {}
    for row in data:
        polyline_id, direction, x, y = int(row[0]), row[1], int(row[2]), int(row[3])
        if polyline_id not in polylines:
            polylines[polyline_id] = []
        polylines[polyline_id].append((x, y))
    return polylines

def thinning(src):
    # Ensure the source is binary
    src = (src == 255).astype(np.uint8) * 255
    # Apply OpenCV's thinning algorithm
    thinned = cv2.ximgproc.thinning(src)
    return thinned

def process_and_classify(csv_path):
    polylines = read_csv(csv_path)
    img = np.zeros((500, 500), dtype=np.uint8)

    for points in polylines.values():
        points = np.array(points, dtype=np.int32)
        cv2.polylines(img, [points], isClosed=False, color=255, thickness=1)

    print("Original Doodles:")
    plt.imshow(img, cmap='gray')
    plt.title("Original Doodles")
    plt.show()

    # Invert colors for better contour detection
    bin_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]

    print("Binary Image:")
    plt.imshow(bin_img, cmap='gray')
    plt.title("Binary Image")
    plt.show()

    thinned_img = thinning(bin_img)

    print("Thinned Image:")
    plt.imshow(thinned_img, cmap='gray')
    plt.title("Thinned Image")
    plt.show()

    # Finding contours in the thinned image
    contours, _ = cv2.findContours(thinned_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    print(f"Number of contours found: {len(contours)}")

    output_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    rng = np.random.default_rng(123)

    for contour in contours:
        hull = cv2.convexHull(contour)
        area = cv2.contourArea(hull)
        perimeter = cv2.arcLength(hull, True)
        circularity = (4 * np.pi * area) / (perimeter * perimeter)

        color = rng.integers(0, 255, size=3).tolist()
        if circularity > 0.8:
            # Circle or ellipse
            if len(hull) >= 5:
                print("Detected: Ellipse")
                ellipse = cv2.fitEllipse(hull)
                cv2.ellipse(output_img, ellipse, color, 2)
            else:
                print("Detected: Circle")
                center, radius = cv2.minEnclosingCircle(hull)
                cv2.circle(output_img, (int(center[0]), int(center[1])), int(radius), color, 2)
        else:
            approx = cv2.approxPolyDP(contour, 0.02 * perimeter, True)
            if len(approx) == 3:
                print("Detected: Triangle")
                # Triangle
                cv2.drawContours(output_img, [approx], 0, color, 2)
            elif len(approx) == 4:
                # Rectangle or square
                x, y, w, h = cv2.boundingRect(approx)
                aspect_ratio = w / float(h)
                if 0.9 <= aspect_ratio <= 1.1:
                    print("Detected: Square")
                    cv2.rectangle(output_img, (x, y), (x + w, y + h), color, 2)  # Square
                else:
                    print("Detected: Rectangle")
                    cv2.rectangle(output_img, (x, y), (x + w, y + h), color, 2)  # Rectangle
            else:
                # Other shapes
                print("Detected: Other shape")
                cv2.drawContours(output_img, [approx], 0, color, 2)

    return output_img

def plot_image(img, title):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

csv_path = "/content/isolated.csv"
output_img = process_and_classify(csv_path)
plot_image(output_img, 'Processed Image')
```
